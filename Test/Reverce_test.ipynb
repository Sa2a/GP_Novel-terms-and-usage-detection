## Prepare Test Data

```python
import pandas as pd
import ast

def to_dict(x):
    try:
        y = ast.literal_eval(x)
        if type(y) == dict:
            return y
    except:
        return None
test_data = pd.read_excel("../Spelling  Correction/Data after Correction/Nove_Misspelled_test.xlsx")
test_data
```

```python

test_data.correct = test_data.correct.apply(to_dict)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
test_data.Label = encoder.fit_transform(test_data.Label)
test_data
```

# Model Scanner Labels 

```python
scan_report =[]
for i, sentence in enumerate(np.uniqu(test_data.sentence.values)):
    #  get all the prediction of the word in the sentence even when it is repeated
    scanned = []
    for word in sentence.split():
        if word not in scanned:
            pred = getRelevanceScore(sentence, word)
            scanned.append(word)
            for p in pred:
                scan_report.append({sentence, word ,p[0], p[1], p[2], p[3]})
```
