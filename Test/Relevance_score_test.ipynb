## Prepare Test Data

```python
import pandas as pd
import ast

def to_dict(x):
    try:
        y = ast.literal_eval(x)
        if type(y) == dict:
            return y
    except:
        return None
test_data = pd.read_excel("../Spelling  Correction/Data after Correction/Nove_Misspelled_test.xlsx")
test_data
```

```python

test_data.correct = test_data.correct.apply(to_dict)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
test_data.Label = encoder.fit_transform(test_data.Label)
test_data
```

## Test

```python
test_report=[]
for i, sentence in enumerate(test_data.sentence.values):
    word = test_data.new_word[i]
    if len(test_data.correct.iloc[i].keys()) == 0 or test_data.correct.iloc[i][word][0] == "لم يتم العثور علي إقتراحات مناسبة لهذه الكلمة":
        corrected_word  = word
        sentence_test = re.sub(r'\b'+word+r"\b",corrected_word,sentence)
        pred = getRelevanceScore(sentence_test, corrected_word)
        test_report.append({"corrected":corrected_word, "prediction": pred})
    else:
        prediction_each_corrected =[]
        for corrected_word in test_data.correct.iloc[i][word]:
            sentence_test = re.sub(r'\b'+word+r"\b",corrected_word,sentence)
            pred = getRelevanceScore(sentence_test, corrected_word)
            prediction_each_corrected.append({"corrected": corrected_word, "prediction": pred})

        # get the most reelvant correction 
        most_relevant  = prediction_each_corrected[-1]
        for predicted in prediction_each_corrected:
            # print(predicted)
            if predicted["prediction"][0][1] > most_relevant["prediction"][0][1]:
                    most_relevant = predicted
        test_report.append(most_relevant)
```

```python
test_report_df = pd.DataFrame(test_report)
test_report_df
```

```python
test_report_df["Label"] = test_report_df.prediction.apply(lambda x: x[0][0])
```

```python
test_report_df
```

```python
test_data
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(test_data.Label, test_report_df.Label)
```

```python
from ar_corrector.corrector import Corrector
corr = Corrector()

corr.spell_correct('ولا نايم')
```

```python
tokenizer.tokenize("ماتجبرش")
```

```python
S = "ومتقوليش متفرقش عشان تفرق طبعا"
corr.contextual_correct(S)
```

```python
def corrector(word):
    correct = corr.spell_correct(word, 1)
    return word if correct == True else correct[0][0]
test_data["ar-corrector"] = test_data.new_word.apply(corrector)

test_data
```

```python
new_test_data = test_data[test_data.Column1 >0]
new_test_data.reset_index(inplace = True,drop=True)
```

```python
test_report2=[]
for i, sentence in enumerate(new_test_data.sentence.values):
    word = new_test_data.new_word[i]
    corrected_word  = new_test_data["ar-corrector"].iloc[i]
    sentence_test = re.sub(r'\b'+word+r"\b",corrected_word,sentence)
    pred = getRelevanceScore(sentence, word)
    test_report2.append({"corrected":corrected_word, "prediction": pred})
    
```
