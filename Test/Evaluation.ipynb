```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("./taghreed_report.xlsx")
```

```python
data
```

```python
conditions = [
    (data['corrected'] == "True") & (data["relevance_no_correction"] <= 0.01),
    (data['corrected'] == "True") & (data['relevance_no_correction'] > 0.8),
    (data['corrected'] != "True") & (data[["relevance_no_correction","relevance_with_correction"]].max(axis=1) >= 0.8),
    (data['corrected'] != "True") & (data[["relevance_no_correction","relevance_with_correction"]].max(axis=1) < 0.1)
    ]

# create a list of the values we want to assign for each condition
# 0 for other conditions
values = ['Change', 'No Change', 'Misspelled', 'Novel']

# create a new column and use np.select to assign values to it using our lists as arguments
data['Label'] = np.select(conditions, values)
```

```python
def applyFilter(novelTh,meaningTh):
    # global originalData
    # data  = originalData.copy()
    conditions = [
        (data['corrected'] == True) & (data["relevance_no_correction"] <= meaningTh),
        (data['corrected'] == True) & (data["relevance_no_correction"] > meaningTh),
        (data['corrected'] != True) & (data[["relevance_no_correction","relevance_with_correction"]].max(axis=1) <= novelTh),
        (data['corrected'] != True) & (data[["relevance_no_correction","relevance_with_correction"]].max(axis=1) > novelTh)
        ]
    values = ['Meaning Change', 'Meaning No Change', 'Novel', 'Misspelled']
    # create a new column and use np.select to assign values to it using our lists as arguments
    data['Label'] = np.select(conditions, values)
    return data
data = applyFilter(novelTh = .5, meaningTh =.5)
data
```
