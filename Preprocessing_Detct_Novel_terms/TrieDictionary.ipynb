# Read the data string and convert it to a set

```python
dictionary_string =[]
with open('/kaggle/working/all_words_train.txt', 'r') as f:
    dictionary_string = f.read()
```

```python
dictionary_set  = set(list(word.strip("'") for word in dictionary_string.strip("[|]").split(", ")))
```

## Test the trie

```python
from marisa_trie import Trie
# trie = Trie([u'key1', u'key2', u'key12'])
trie = Trie(set(["خالد", "خالد محمد", "السقا","انا", "انا بحبك"]))

def exists(trie, word):
    return word in trie
assert exists(trie, "خالد")
assert exists(trie, "خالد محمد")
assert exists(trie, "السقا")
assert not exists(trie, "")
assert not exists(trie, "خا")
# trie = Trie([], trie)
assert exists(trie, "انا")
assert exists(trie, "خالد")
assert exists(trie, "انا بحبك")
assert exists(trie, "السقا")
assert not exists(trie, "")
assert not exists(trie, "خال")
assert not exists(trie, "hello ")
assert not exists(trie, "hello ag")
assert not exists(trie, "byebye")
```

## Create Trie dictionary for the MARBert unique words
